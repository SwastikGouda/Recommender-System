Popularity Based Recommendation Systems

```python
ratings_with_name=ratings.merge(book,on='ISBN')
```

```python
ratings_with_name.head(4)
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'Num-Rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name[['Book-Title','Book-Rating']].groupby('Book-Title').mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['Num-Rating']>=250].sort_values('avg_rating',ascending=False)
```

```python
popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num-Rating','avg_rating']].reset_index(drop=True)
```

```python
## Collaborative
```

```python
x=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
x=x[x['Book-Rating']>200]
x
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(x['User-ID'])]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
#!pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
np.where(pt.index=='Zoya')[0][0]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend("Harry Potter and the Chamber of Secrets (Book 2)")
```
